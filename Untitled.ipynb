```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
MIN_MATCH_COUNT = 10
img1 = cv.imread('Screen/current.jpg', cv.IMREAD_GRAYSCALE)          # queryImage
img2 = cv.imread('Screen/previous.jpg', cv.IMREAD_GRAYSCALE)         # trainImage
img1 = cv.Canny(img1,300,500)
img2 = cv.Canny(img2,300,500)
# Initiate SIFT detector
sift = cv.SIFT_create()
# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)
FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks = 50)
flann = cv.FlannBasedMatcher(index_params, search_params)
matches = flann.knnMatch(des1,des2,k=2)
# store all the good matches as per Lowe's ratio test.
good = []
for m,n in matches:
    if m.distance < 1*n.distance:
        good.append(m)
```

```python
src_pts = None
dst_pts = None
if len(good)>MIN_MATCH_COUNT:
    src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
    dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)
    M, mask = cv.findHomography(src_pts, dst_pts, cv.RANSAC,5.0)
    matchesMask = mask.ravel().tolist()
    h,w = img1.shape
    pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
    dst = cv.perspectiveTransform(pts,M)
    img2 = cv.polylines(img2,[np.int32(dst)],True,255,3, cv.LINE_AA)
else:
    print( "Not enough matches are found - {}/{}".format(len(good), MIN_MATCH_COUNT) )
    matchesMask = None
```

```python
draw_params = dict(matchColor = (0,255,0), # draw matches in green color
                   singlePointColor = None,
                   matchesMask = matchesMask, # draw only inliers
                   flags = 2)
img3 = cv.drawMatches(img1,kp1,img2,kp2,good,None,**draw_params)
plt.imshow(img3, 'gray'),plt.show()
```

```python
src_array = src_pts.reshape(-1, 1)
dst_array = dst_pts.reshape(-1)
length = src_array.shape[0] // 2
extra1 = np.array([[1, 0]]).repeat(length, axis=0).reshape(-1, 1)
extra2 = np.array([[0, 1]]).repeat(length, axis=0).reshape(-1, 1)
equM = np.hstack((src_array, extra1, extra2))
# matrix multiplication
M = np.matmul(np.transpose(equM), equM)
X = np.matmul(np.transpose(equM), dst_array)
# solve the equation
a, b, c = np.linalg.solve(M, X)
print(a, b, c)
```

```python
draw_params = dict(matchColor = (0,255,0), # draw matches in green color
                   singlePointColor = None,
                   flags = 2)
img3 = cv.drawMatches(img1,kp1,img2,kp2,good,None,**draw_params)
plt.imshow(img3, 'gray'),plt.show()
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img1 = cv.imread('Screen/previous.jpg', cv.IMREAD_GRAYSCALE)
img2 = cv.imread('Screen/current.jpg', cv.IMREAD_GRAYSCALE)
assert img1 is not None, "file1 could not be read, check with os.path.exists()"
assert img2 is not None, "file2 could not be read, check with os.path.exists()"
edges1 = cv.Canny(img1,300,500)
edges2 = cv.Canny(img2,300,500)
plt.subplot(221),plt.imshow(img1,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(edges1,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(img2,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(edges2,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```

```python
valueMax = 0
size = edges1.shape
for i in range(41):
    scale = 0.9 + i * 0.005
    test1 = edges1.astype(np.float64)
    test2 = edges2.astype(np.float64)
    shift = (1 - scale) / 2 * size[0]
    test1 = cv.warpAffine(test1, np.float32([[scale,0,shift],[0,scale,shift]]),(size[0],size[1]))
    test2 = cv.warpAffine(test2, np.float32([[1,0,0],[0,1,0]]),(size[0],size[1]))
    offset, value = cv.phaseCorrelate(test1, test2)
    print(offset, value, scale)
    if value > valueMax:
        valueMax = value
        final1 = test1
        final2 = test2
        res = cv.warpAffine(test1, np.float32([[1,0,offset[0]],[0,1,offset[1]]]),(size[0],size[1]))
plt.subplot(131),plt.imshow(final1,cmap = 'gray')
plt.subplot(132),plt.imshow(final2,cmap = 'gray')
plt.subplot(133),plt.imshow(res - final2,cmap = 'gray')
print(valueMax)
```
